```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv('billboard.csv')
```

```python
# Let's explore

df.head()
```

```python
print(df.isnull().sum())
```

part A

```python
# Group the data by song and performer, and sum the weeks spent on the chart

song_weeks = df.groupby(['performer', 'song'])['week_position'].count().reset_index()
song_weeks = song_weeks.rename(columns={'week_position': 'count'})

# Sort the data by count in descending order and get the top 10
top_songs = song_weeks.sort_values(by='count', ascending=False).head(10)

# Display the top songs in a table format

print("Caption: Top 10 Most Popular Songs and performers on Billboard Top 100")
print(top_songs[['performer', 'song', 'count']].to_string(index=False))
```

part B

```python
# Filter out the years 1958 and 2021
df_filtered = df[(df['year'] != 1958) & (df['year'] != 2021)]
```

```python
# Group by year and calculate the number of unique songs for each year
diversity_data = df_filtered.groupby('year')['song'].nunique()

print(diversity_data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(diversity_data.index, diversity_data.values, marker='o')
plt.title('Musical Diversity in Billboard Top 100 (1959-2020)')
plt.xlabel('Year')
plt.ylabel('Number of Unique Songs')
plt.grid(True)
plt.show()
```

Part C

```python
# Group by artist and song, and count the number of weeks each song was on the chart
artist_song_weeks = df_filtered.groupby(['performer', 'song'])['week_position'].count().reset_index()


# Filter songs that appeared for at least ten weeks
ten_week_hits = artist_song_weeks[artist_song_weeks['week_position'] >= 10]

print(ten_week_hits)
```

```python
# Count the number of ten-week hits for each artist
artist_ten_week_hits = ten_week_hits['performer'].value_counts()

# Filter out artists with fewer than 30 ten-week hits
popular_artists = artist_ten_week_hits[artist_ten_week_hits >= 30]

print(popular_artists)
```

```python
# Create the bar plot
plt.figure(figsize=(10, 6))
popular_artists.plot(kind='bar')
plt.title('Top Artists with at Least 30 "Ten-Week Hits" on Billboard Top 100')
plt.xlabel('Artist')
plt.ylabel('Number of Ten-Week Hits')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Caption:
Bar plot displaying the number of "Ten-Week Hits" for the top 19 artists in U.S. musical history since 1958 to 2020. A "Ten-Week Hit" is defined as a single song that appeared on the Billboard Top 100 for at least ten weeks. 

Commentary:
The bar plot showcases the remarkable careers of several iconic artists who have consistently produced chart-topping hits that remained on the Billboard Top 100 for at least ten weeks. Elton John stands out at the top of the list with an impressive 52 "Ten-Week Hits," showing his influence on the music industry.

Madonna follows closely with 44 "Ten-Week Hits," demonstrating her consistent ability to capture the attention of listeners over the decades. Meanwhile, country music stars Kenny Chesney and Tim McGraw have secured their places in music history with 41 and 39 "Ten-Week Hits," respectively, highlighting the enduring popularity of country music on the charts.
